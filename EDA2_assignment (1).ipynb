```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# 1. Data Exploration and Preprocessing

# Basic Data Exploration
print(data.head())
print(data.info())
print(data.describe())

# Missing Values
print(data.isnull().sum())

# Handling Missing Values
# Replace '?' with NaN
data.replace('?', np.nan, inplace=True)

# For numerical features, we can impute with the mean
for column in data.columns:
  if data[column].dtype in ['int64', 'float64']:
    data[column].fillna(data[column].mean(), inplace=True)


# For categorical features, we can impute with the mode
for column in data.columns:
  if data[column].dtype == 'object':
    data[column].fillna(data[column].mode()[0], inplace=True)

print("Missing values after handling:\n", data.isnull().sum())
```

```python
numerical_features = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']

# 1. Standard Scaling
scaler = StandardScaler()
data_standard_scaled = data.copy()
data_standard_scaled[numerical_features] = scaler.fit_transform(data[numerical_features])
data_standard_scaled.head()

# 2. Min-Max Scaling
scaler = MinMaxScaler()
data_minmax_scaled = data.copy()
data_minmax_scaled[numerical_features] = scaler.fit_transform(data[numerical_features])
data_minmax_scaled.head()


print("\nData with Standard Scaling:")
print(data_standard_scaled.head())

print("\nData with Min-Max Scaling:")
print(data_minmax_scaled.head())

```

```python
# Discussion on Scaling Techniques
# Standard Scaling:
# * Preferred when the data follows a normal distribution or when we want to
#   transform the data to have a mean of 0 and a standard deviation of 1.
# * It is sensitive to outliers.

# Min-Max Scaling:
# * Preferred when the data does not follow a normal distribution, or when we
#   want to scale the data to a specific range (e.g., 0 to 1).
# * It is less sensitive to outliers compared to Standard Scaling.
```

2. Encoding Techniques:

*Apply One-Hot Encoding to categorical variables with less than 5 categories.

*Use Label Encoding for categorical variables with more than 5 categories.

*Discuss the pros and cons of One-Hot Encoding and Label Encoding.


```python
# Identify categorical features
categorical_features = [col for col in data.columns if data[col].dtype == 'object']

# One-Hot Encoding for features with less than 5 categories
for feature in categorical_features:
  if len(data[feature].unique()) < 5:
    data = pd.get_dummies(data, columns=[feature], prefix=[feature])
```

```python
# Label Encoding for features with more than 5 categories
from sklearn.preprocessing import LabelEncoder

categorical_features = [col for col in data.columns if data[col].dtype == 'object']

for feature in categorical_features:
  if len(data[feature].unique()) >= 5:
    le = LabelEncoder()
    data[feature] = le.fit_transform(data[feature])

print("\nData after encoding:")
print(data.head())
```

```python
# Discussion on Encoding Techniques

# One-Hot Encoding:
# Pros:
# * Creates a binary representation for each category, which can be easily used by
#   machine learning algorithms.
# * Avoids introducing an artificial order to the categories.
# Cons:
# * Can lead to a high number of features if there are many categories in a
#   variable.
# * Can make the dataset sparse, which can be computationally expensive.

# Label Encoding:
# Pros:
# * Reduces the number of features compared to One-Hot Encoding.
# * Can be useful when the categorical variable has a natural order (e.g., low,
#   medium, high).
# Cons:
# * Introduces an artificial order to the categories if there is no inherent
#   order.
# * Can mislead machine learning algorithms into assuming a relationship between
#   the encoded values that may not exist.
```

3. Feature Engineering:

*Create at least 2 new features that could be beneficial for the model.

*Explain the rationale behind your choices.

*Apply a transformation (e.g., log transformation) to at least one skewed numerical feature and justify your choice.


```python
# Create new features
data['capital_gain_loss'] = data['capital_gain'] - data['capital_loss']
data['age_education_interaction'] = data['age'] * data['education_num']
```

```python
# Rationale:
# * capital_gain_loss: Combining capital gain and loss into a single feature
#   might capture the overall financial status better.
# * age_education_interaction: This feature captures the interaction between age
#   and education level, potentially indicating a person's experience and
#   knowledge.
```

```python
# Log Transformation for a skewed feature
# We can apply log transformation to 'fnlwgt' (if it's significantly skewed)
# to make its distribution more normal.

if data['fnlwgt'].skew() > 1:  # Check for significant skewness
  data['fnlwgt_log'] = np.log1p(data['fnlwgt'])

print("\nData after feature engineering:")
print(data.head())
```

```python
# Justification for log transformation:
# Log transformation is applied to reduce the impact of outliers and make the
# distribution of the feature more normal. This can improve the performance of
# some machine learning algorithms that assume a normal distribution.
```

4. Feature Selection:
*Use the Isolation Forest algorithm to identify and remove outliers. Discuss how outliers can affect model performance.

*Apply the PPS (Predictive Power Score) to find and discuss the relationships between features. Compare its findings with the correlation matrix.


```python
# Outlier Detection and Removal using Isolation Forest
from sklearn.ensemble import IsolationForest

# Assuming 'income' is your target variable
X = data.drop('income', axis=1)
y = data['income']

# Perform one-hot encoding on categorical features
categorical_features = ['workclass', 'education', 'marital_status', 'occupation',
                        'relationship', 'race', 'sex', 'native_country']
X_encoded = pd.get_dummies(X, columns=categorical_features)

# Initialize Isolation Forest
iso_forest = IsolationForest(contamination='auto', random_state=42)

# Fit and predict outliers
outliers = iso_forest.fit_predict(X_encoded)

# Remove outliers
data_no_outliers = data[outliers == 1]

print("\nData after outlier removal:")
print(data_no_outliers.head())
```

```python
# Discussion on outliers:
# Outliers can significantly affect the performance of machine learning models
# by:
# * Increasing the variance in the data, leading to less reliable model
#   coefficients and predictions.
# * Misleading model training, as models can be overly sensitive to extreme
#   values.
# * Reducing the accuracy and generalizability of the model.
```
